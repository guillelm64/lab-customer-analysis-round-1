```python
#Read the three files into python as dataframes
import pandas as pd
file1 = pd.read_csv('file1.csv')
file2 = pd.read_csv('file2.csv')
file3 = pd.read_csv('file3.csv')

df1 = pd.DataFrame(file1)
df2 = pd.DataFrame(file2)
df3 = pd.DataFrame(file3)
```

```python
#Dataframes shape

df1.shape
df2.shape
df3.shape
```

```python
#Standardize header names
def make_lower(df):
    columns = df.columns
    col = []
    for colum in df.columns:
        col.append(colum.lower())
    return col
df1.columns = make_lower(df1)
df2.columns = make_lower(df2)
df3.columns = make_lower(df3)

display(df3)
```

```python
#Rearrange the columns in the dataframe as needed
def rearr(df):
    df = df.reset_index(drop = True)
    return df
rearr(df1)
rearr(df2)
rearr(df3)
```

```python
#Concatenate the three dataframes
df_all = pd.concat([df1,df2,df3], axis = 0)
display(df_all.head(15))
```

```python
#Which columns are numerical? 
#Which columns are categorical?
num = df_all.select_dtypes(include = ["int16", "int32", "int64", "float16", "float32", "float64"])
display(num.head())
obj = df_all.select_dtypes(include = ['object'])
display(obj.head())
```
